```python
 # 1mb = 1024 kb
 # so lets write a function to replace M with empty string and then convert it into float and then multiply it with 1024 to convert it into kb, and if the value is in kb then we will just convert it into float and return it, and if the value is Varies with device then we will return NaN. 
def size_process(item):
     if str(item)[-1] == "M":
        res = float(str(item).replace("M", ""))
        res = res*1024
        return res
     elif str(item)[-1] == "k":
        res = float(str(item).replace("k", ""))
        return res
     else:
        return str(np.nan)
         
```

```python
size_process("19M") # to check the function is working or not, it should return 19456.0 because 19*1024=19456.0
```

```python
# top most reviewed apps by category 
top_5_reviewed_apps_by_category = (
    df_copy
    .sort_values(["Category", "Reviews"], ascending=[True, False])
    .groupby("Category").head(5)
    [["Category", "App", "Reviews"]]
    .reset_index(drop=True)
)

top_5_reviewed_apps_by_category.head(5)
```

```python
#groupby category and app, aggregated sum on reviews
top_reviewed_apps_by_category = (
    df_copy.groupby(["Category", "App"])["Reviews"].sum().sort_values(ascending=False).reset_index()
)
top_reviewed_apps_by_category.head(5)
```

```python
sns.barplot(x=top_reviewed_apps_by_category["App"].head(5), y=top_reviewed_apps_by_category["Reviews"].head(5), palette="Set1")
plt.xticks(rotation = 80)
plt.xlabel("App")
plt.ylabel("Reviews")
plt.title("Top Reviewed Apps by Category")  
```
